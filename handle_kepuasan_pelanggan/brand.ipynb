### Brand

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
brands = ['erigo', 'robot', 'sandisk', 'teamgroup', 'gm', 'rinnai', 'xiaomi', 'bardi']
data_dummy = []


brand_ranges = {
    'erigo': (80, 100),
    'robot': (30, 50),
    'sandisk': (60, 80),
    'teamgroup': (40, 60),
    'gm': (50, 70),
    'rinnai': (70, 90),
    'xiaomi': (20, 40),
    'bardi': (10, 30)
}

# Buat DataFrame dengan data dummy untuk setiap merek dan rentang nilai
for brand in brands:
    for _ in range(300):
        data_dummy.append({
            'Brand Recognition': brand,
            'Perubahan': random.randint(brand_ranges[brand][0], brand_ranges[brand][1])
        })

df_brand = pd.DataFrame(data_dummy)
```

```python
df_brand
```

```python
grouped = df_brand.groupby('Brand Recognition')['Perubahan'].mean()

# Visualisasikan dengan grafik bar chart menggunakan Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(grouped.index, grouped.values)
plt.xlabel('Merek')
plt.ylabel('Rata-rata Perubahan')
plt.title('Peningkatan Branding Platform E-commerce')
plt.xticks(rotation=45)
plt.show()
```
