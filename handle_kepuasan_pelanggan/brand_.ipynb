```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Load the data
df_orderitems = pd.read_csv("./dataset/OrderItems-2023-11-14.csv")

# Define brands to match in the product names
brands = ['Erigo', 'ROBOT', 'SANDISK', 'TEAMGROUP', 'GM', 'Rinnai', 'Xiaomi', 'BARDI']


def assign_brand(row):
    for brand in brands:
        if re.search(brand, row['name'], re.IGNORECASE):
            return brand
    return 'Other'

# Add a 'brand' column based on name similarity
df_orderitems['brand'] = df_orderitems.apply(assign_brand, axis=1)



# Create a bar chart
plt.figure(figsize=(10, 6))
df_orderitems['brand'].value_counts().plot(kind='bar')
plt.xlabel('Brand')
plt.ylabel('Jumlah Produk')
plt.title('Jumlah Produk per Brand')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
df_orderitems.to_csv("./dataset/OrderItems_with_brand.csv", index=False)
```
